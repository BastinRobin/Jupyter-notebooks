```python
# Data manipulation and Data cleaning
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('countries.csv')
```

```python
x = [10, 20, 30]
y = [10, 12, 25]
z = [10, 15, 30]
plt.plot(x, y)
plt.plot(x, z)
plt.legend(["Y axis", "Z axis"])
plt.xlabel('X axis')
plt.ylabel('Y and Z axis')
plt.show()
```

```python
# Print all columns of dataframe
data.columns
```

```python
# Select only afghanistan
afg = data[data['country'] == 'Afghanistan']

# Select only India
india = data[data['country'] == 'India']

# Select only China
china = data[data['country'] == 'China']
```

```python
plt.plot(afg.year, afg.population / afg.population.iloc[0] * 100)
plt.plot(india.year, india.population / india.population.iloc[0] * 100)
# plt.plot(china.year, china.population / china.population.iloc[0] * 100)
plt.legend(["Afghanistan", "India", "China"])
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```

```python
plt.plot(india.year, india.population / india.population.iloc[0] * 100)
plt.plot(china.year, china.population / china.population.iloc[0] * 100)
plt.legend(["India", "China"])
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```
